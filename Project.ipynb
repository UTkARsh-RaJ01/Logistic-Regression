```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
d=pd.read_csv('test_data.csv')
```

```python
d.head()
```

```python
X_test=d[['PassengerId','Sex','Age','Fare','Pclass_1','Pclass_2','Pclass_3','Family_size','Title_1','Title_2','Title_3','Title_4','Emb_1','Emb_2','Emb_3']]
y_test=d['Survived']
```

```python
lr.fit(X,y)
```

```python
predict=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
test_data_accuracy = accuracy_score(y_test,predict)
print('Accuracy score of test data : ', test_data_accuracy)
```
